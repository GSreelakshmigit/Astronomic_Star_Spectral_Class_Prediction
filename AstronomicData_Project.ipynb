# K-Nearest Neighbor(KNN) Algorithm

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=0)
```

```python
print("x_train before scaling..")
df3=pd.DataFrame(x_train)
print(df3.to_string())
```

```python
#feature Scaling
from sklearn.preprocessing import StandardScaler
model3= StandardScaler()
x_train= model3.fit_transform(x_train)
x_test= model3.transform(x_test)
```

```python
print("x_train after scaling...")
df4=pd.DataFrame(x_train)
print(df4.to_string())
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )
classifier.fit(x_train, y_train)
```

```python
pred_xtest=classifier.predict(x_test)
print(pred_xtest)
```

```python
print("Prediction comparison")
ddf=pd.DataFrame({"Y_test":y_test,"Y-pred":pred_xtest})
print(ddf.to_string())
```

```python
from sklearn.metrics import accuracy_score
accuracy3 = accuracy_score(y_test, pred_xtest)
print('Accuracy: %.2f' % (accuracy3*100))
```

# Support Vector Machine Algorithm


```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear', random_state=0)
classifier.fit(x_train, y_train)
```

```python
y_pred= classifier.predict(x_test)
```

```python
df2=pd.DataFrame({"Actual Y_Test":y_test,"Prediction Data":y_pred})
print("prediction status")
print(df2.to_string())
```

```python
accuracy4 = accuracy_score(y_test, y_pred)
print('Accuracy: %.2f' % (accuracy4*100))
```

# Random Forest Algorithm 


```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=10,criterion="entropy")
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
print("------------PREDICTION----------")
df2=pd.DataFrame({"Actual Result-y":y_test,"Prediction Result":y_pred})
print(df2.to_string())
```

```python
from sklearn import metrics
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,y_pred))
print('Mean Squared Error:',metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
#evaluate prediction
from sklearn.metrics import accuracy_score
accuracy5=accuracy_score(y_test,y_pred)
print('Accuracy:%.2f'%(accuracy5*100))
```

# Naïve Bayes Classifier Algorithm

```python
y_pred=classifier.predict(x_test)
print("------------PREDICTION----------")
df2=pd.DataFrame({"Actual Result-Y":y_test,"Prediction Result":y_pred})
print(df2.to_string())
```

```python
from sklearn import metrics
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,y_pred))
print('Mean squared Error:',metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
accuracy6=accuracy_score(y_test,y_pred)
print('Accuracy:%.2f'%(accuracy6*100))
```
